# **DATA DESCIPTION**

EmployeeID:

Description: Unique identifier assigned to each employee.
Type: Integer
Age:

Description: Age of the employee in years.
Type: Integer
Attrition:

Description: Indicates whether the employee has left the company (Yes/No).
Type: Categorical (Yes/No)
BusinessTravel:

Description: Frequency of business travel for the employee.
Type: Categorical (Travel_Rarely, Travel_Frequently, Non-Travel)
Department:

Description: Department where the employee works.
Type: Categorical (Sales, Research & Development, Human Resources)
DistanceFromHome:

Description: Distance between the employees home and workplace (in kilometers).
Type: Integer
Education:

Description: Education level of the employee.
Type: Integer (1 = 'Below College', 2 = 'College', 3 = 'Bachelor', 4 = 'Master', 5 = 'Doctor')
EducationField:

Description: Field of education of the employee.
Type: Categorical (Life Sciences, Medical, Marketing, Technical Degree, Other)
EmployeeCount:

Description: Total number of employees in the company.
Type: Integer (usually constant, e.g., 1 for each row)
Gender:

Description: Gender of the employee.
Type: Categorical (Male, Female)
JobLevel:

Description: Job level of the employee within the organization.
Type: Integer
JobRole:

Description: Specific job role of the employee.
Type: Categorical (e.g., Sales Executive, Research Scientist)
MaritalStatus:

Description: Marital status of the employee.
Type: Categorical (Single, Married, Divorced)
MonthlyIncome:

Description: Monthly income of the employee.
Type: Integer
NumCompaniesWorked:

Description: Number of companies the employee has worked for before the current job.
Type: Integer
Over18:

Description: Whether the employee is over 18 years old (usually constant, e.g., 'Yes').
Type: Categorical
PercentSalaryHike:

Description: Percentage increase in salary over the previous year.
Type: Integer
StandardHours:

Description: Standard working hours (usually constant, e.g., 80).
Type: Integer
StockOptionLevel:

Description: Stock option level of the employee.
Type: Integer
TotalWorkingYears:

Description: Total number of years the employee has been working.
Type: Integer
TrainingTimesLastYear:

Description: Number of training sessions attended by the employee last year.
Type: Integer
YearsAtCompany:

Description: Number of years the employee has been with the company.
Type: Integer
YearsSinceLastPromotion:

Description: Number of years since the employees last promotion.
Type: Integer
YearsWithCurrManager:

Description: Number of years the employee has been working with the current manager.
Type: Integer
EnvironmentSatisfaction:

Description: Satisfaction with the work environment (1 = Low, 2 = Medium, 3 = High, 4 = Very High).
Type: Integer
JobSatisfaction:

Description: Satisfaction with the job (1 = Low, 2 = Medium, 3 = High, 4 = Very High).
Type: Integer
WorkLifeBalance:

Description: Work-life balance rating (1 = Bad, 2 = Good, 3 = Better, 4 = Best).
Type: Integer
JobInvolvement:

Description: Level of involvement in the job (1 = Low, 2 = Medium, 3 = High, 4 = Very High).
Type: Integer
PerformanceRating:

Description: Performance rating of the employee (1 = Low, 2 = Good, 3 = Excellent, 4 = Outstanding).
Type: Integer

# **FEATURE ENGINEERING**

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Gender'] = encoder.fit_transform(df["Gender"])
df['Attrition'] = encoder.fit_transform(df["Attrition"])
df['BusinessTravel'] = encoder.fit_transform(df["BusinessTravel"])
df['Department'] = encoder.fit_transform(df["Department"])
df['EducationField'] = encoder.fit_transform(df["EducationField"])
df['JobRole'] = encoder.fit_transform(df["JobRole"])
df['MaritalStatus'] = encoder.fit_transform(df["MaritalStatus"])
df['Over18'] = encoder.fit_transform(df["Over18"])
df
```
